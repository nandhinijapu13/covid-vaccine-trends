```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\nandh\\Downloads\\owid-covid-data.csv.csv")  
us = df[df['location'] == 'United States']

plt.figure(figsize=(12,6))
plt.plot(us['date'], us['new_cases_smoothed'], label='New Cases')
plt.plot(us['date'], us['people_fully_vaccinated'], label='Fully Vaccinated')
plt.title('COVID-19 Trends in the U.S.')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
